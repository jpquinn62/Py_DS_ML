```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
# lmplot=linear model, scatter plot w/ linear fit, add hue for seperation
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v']) 
# add in matplotlib markers to affect the plot
# the 'o' and 'v' are symbols that comes from a matplotlib library so look them up 
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
# note curly brace for dictionary using colon ':', and a call to matplotlib as to whats going on underhood of seaborn
# if too small to see, add this in, pass in a dictionary, and made the marker a lot larger
```

```python
# can customize by using grid instead of using hue=''
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex') # now column plots two grids based on the variable
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time') # note it breaks out to 4 grids
```

```python
# can add hue back in and switch some cols names around
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,col="day",hue="sex"
          ,aspect=0.6,height=8) #size=8) size is now aka height to avoid user warning 
# nice size and ascpet look wrong, so... change with ascpet and size
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm') # easier on the eyes
```
