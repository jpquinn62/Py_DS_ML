```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) # defaults to median, # avg/mean total bill per category of 'sex' # shift-tab to see the formual inputs
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std) # add estimator and standard deviation , but first import numpy
```

```python
# above, x is a categorical column, and y is a numerical column with barplot, it's an aggregate or group by mechanism
```

```python
sns.countplot(x='sex',data=tips) # much like barplot w/ estimator, but estiamtor explictly counts number of occurances 
```

```python
# Boxplot and violinplot
```

```python
# boxplot aka box and whisker plot
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker') # points above whiskers are outliers 
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True) # violine shows more, but do both box and violine together to see even more visual information
```

```python
#sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True) # scater plot based off of category which is x, because you cant see the plotted pointes on top of ea other, add jitter
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True) # scater plot based off of category which is x, because you cant see the plotted pointes on top of ea other, add jitter
```

```python
# swarm plot, it's like a combo strip plot and violin plot
```

```python
# gives a better represenation of the distribution of values 
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# one drawback, they tend to not scale well when plotting very large datasets 
```

```python
# show in combo with a violin plot, as most people have not seen a swarm plot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```
