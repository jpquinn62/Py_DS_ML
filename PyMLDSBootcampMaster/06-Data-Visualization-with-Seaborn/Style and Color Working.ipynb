```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# setting particular styles 
```

```python
# setting this, will set the style for the rest of your plots
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks') # remove spines
sns.countplot(x='sex',data=tips)
sns.despine() # removes top spines
```

```python
sns.set_style('ticks') # remove spines
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True) # add more as params to remove even more spines
# keeping spines can look better with more guideline info 
```

```python
# scale and context
```

```python
sns.set_context('poster') # change to a poster instead of notebook
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=3) # change font
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('notebook',font_scale=1) # change font and context 
sns.countplot(x='sex',data=tips)
```
